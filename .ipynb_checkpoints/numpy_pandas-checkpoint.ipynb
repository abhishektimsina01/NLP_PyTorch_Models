```python
k = torch.randint(low=0, high= 10, size=(2,3))
l = k.clone()
k
```

```python
torch.sigmoid(k)
```

```python
torch.log(k)
```

```python
torch.sqrt(k)
```

```python
torch.square(k)
```

```python
l = l.to(torch.float32)
op = torch.softmax(l, dim=1)
op
```

```python
for i in op:
    a = i[0] + i[1] + i[2]
print(a)
```

```python
torch.tanh(k)
```

```python
torch.relu(k)
```
