```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import train_test_split
```

```python
full_df = pd.read_csv("fashion-mnist_test.csv")
full_df
```

```python
df = full_df[0:6000]
df.shape
```

```python
features = df.iloc[:, 1:]
features
```

```python
labels = df.iloc[:,0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size= 0.2, random_state=42, shuffle= True)
```

```python
x_train = x_train/255.0
```

```python
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = torch.tensor(features.to_numpy()).to(torch.float32)
        self.labels = torch.tensor(labels.to_numpy()).to(torch.long)
    
    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
```

```python
train_dataset = CustomDataset(x_train, y_train)
```

```python
test_dataset = CustomDataset(x_test, y_test)
```

```python
class ImageClassificationModel(nn.Module):

    def __init__(self, number_features):        
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(number_features, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32,10)
        )
        # optimization/ gradient descent
        self.optim = torch.optim.SGD(self.model.parameters(), lr=0.1)
    
    def forward(self, features):
        self.y_pred = self.model(features)
        return self.y_pred
    
    def loss_function(self, labels):
        loss_fn = nn.CrossEntropyLoss()
        self.loss = loss_fn(self.y_pred, labels)
        return self.loss.item()

    def optimization(self):
        self.optim.zero_grad()
        self.loss.backward()
        self.optim.step()
```

```python
train_dataset.__len__()
```

```python
x_train.shape
```

```python
model = ImageClassificationModel(x_train.shape[1])
```

```python
dataloader = DataLoader(train_dataset, batch_size=32, shuffle= True)
len(dataloader)
```

```python
epochs = 100
```

```python
# mini-batch gradient descent
# perfor optimization on each mini batch
batch = 0
for epoch in range(epochs) :
    total_loss = 0
    for features_batch, target_batch in dataloader:
        y_pred = model(features_batch)
        loss = model.loss_function(target_batch)
        model.optimization() 
        total_loss = total_loss + loss
    avg_loss = total_loss/len(dataloader)
    print("epoch:", epoch+1)
    print("loss", avg_loss)
```
