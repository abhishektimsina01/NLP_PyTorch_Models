```python
# training a deep neural network
```

```python
# one neuron input layer and one neuron output layer (regression)
# one neuron hidden layer (sigmoid)
# forward propagation
# loss function
# backpropagation -> gradient(dL/dW and dL/db) that is the loss due to the wieght and bias w and b, 
# gradient descent is adjusting the wieght towards the opitmal value where the loss due to the weight become nearly 0 (still loss but not because of the wieght)
```

```python
# forward propagtion 
# ==> z = x*w + b
# ==> y = sig(z)
# ==> loss = (y - y(real)^2) (SSR)
```

```python
# backward propagation:- 
##### ==> we need to calculate the gradient or slope dL/dw, 
##### that is the rate with which the loss falls or rises when the weight is increased or decresed.
##### The slope also gives the contribution of weight in bringing the loss L, not all but certain percentage is brought
##### which we try to reduce it.
##### we reduce the dl/dw to nearly 0 which means there is almost no contrubution of w to the loss(L) that is the w is in optimal solution
##### It is done using the gradient descent, but we increase it slowly. The far we are from the minima point then more the slope and
##### the bigger the step and as we move closer to the point of no loss due to the weight then we take less step to avoid overshooting
##### using chain rule to find the relation between the loss and w
```

```python
# the gradient formula will be:
# dL/dw = dL/dy(predicted) * dy(predicted)/dz * dz(sigmoid)/dw
```

```python
# performing the differentiation using the AutoGrad
```

derivative of y = $x^2$

dy/dx = 2x

```python
x = torch.tensor(3.0, requires_grad=True)
z = torch.tensor(3.0, requires_grad=True)

# "requires_grad" makes the pyTorch make the computational graph to remember and
# track the operation it will be used in the future so that we can make the gradient w.r.to this    
```

if we add the requires_grad property as True for a tensor then PyTorch will:

Track all operations done using that tensor

Build a computational graph

Allow you to compute gradients of some final output with respect to that tensor

In ANN or in any Deep Learning models, 
finding the weight's contribution in loss (dL/dw) and 
finding the bias's contribution in loss (dL/db) can be a lot complex as we have to find the gradient for each of the weights on our own if we dont use the the requires_grad.
And if we do, then all the computational graph will be calculated respect to the tensor and we can calcuate the gradient easily.

```python
x = torch.tensor(6.7)
y_real = torch.tensor(0.0)

weight = torch.tensor(1.0, requires_grad= True)
bias = torch.tensor(0.0, requires_grad= True)
```

```python
# weight = torch.rand(1)
# # w = weight[0]
# weight = weight[0]
# weight
```

```python
# lineasr combination for the raw value
z = weight * x + bias
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
y_real
```

```python
# Cross entropy as Loss function:

loss_fn = torch.nn.BCELoss()
loss = loss_fn(y_pred, y_real)
L = loss.item()
L
```

```python
y_pred
```

```python
y_real
```

```python
# gradients
# since the loss has the relation with the y_pred, we use the loss for backward as it is also the actual loss
loss.backward()
dL_dw = weight.grad
dL_db = bias.grad
```

```python
dL_dw
```

```python
dL_db
```

```python
# gradient descent
weight = weight - (dL_dw*0.0001)
bias = bias - (dL_db*0.001)
```
