## Transformers from scratch

##### The transformer is a seq2seq deep learning model, that takes an input sequence(parallelly) and produces the output sequence either one token at a time (inferece) or all at a time parallelly(training time).

##### The transformer acts as the AutoRegressive model(AR model) in the case of the inference time and as non-AutoRegressive in the case of training due to the known output and to reduce the training time.


##### ->There are two blocks in the transformer, encoder and decoder block. There are 6 identical encoder units in the encoder block and 6 identical decoder unit in the decoder block.
##### 1. The main goal of the encoder block is to understand the context and have the deep, rich and refined meaning of the input sentence by understanding the meaning of the each token. It have the deep, transformed meaning/context of each of the token.
##### 2. The main goal of the decoder block is to generate the token based upon the input sentence from the encoder and the tokens genereated so far. It uses the teacher forcing method to perform the training parallelly.
